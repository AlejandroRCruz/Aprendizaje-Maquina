# Problema simple de regresión donde buscamos predecir el precio de una casa según sus caracteristicas-

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df=pd.read_csv("data.csv")
df.head()
```

```python
print(f'hay {len(df)} entradas')
print(f'los datos tienes {len(df.columns)} caracteristicas')
df.info()
```

```python
#calculamos estadisticas importantes 
df.describe()
```

```python
#Vemos si existen null
df.isnull().sum()
```

```python
# vemos el promedio de vivienda según cada ciudad
df.groupby("city")[["price"]].mean()
```

```python
# vemos cuantas casas fueron construidas en cada año
print(df['yr_built'].value_counts())
#Vemos cual es el precio maximo minimo y promedio de las casas
print(f"\nEl valor maximo de una casa fue: {df['price'].max()} en el año: {df['yr_built'][df['price']==df['price'].max()].iloc[0]}") # Si no usamos iloc 
#nos devolvera el indice y el año además del tipo de dato del valor
print(f'\nEl minimo fue {df['price'].min()} en el año  {df['price'].max()} en el año: {df['yr_built'][df['price']==df['price'].min()].iloc[0]}')
print(f'\nEl promedio fue {df['price'].mean()}')
```
